```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'Rating-Avg'},inplace = True)
```

```python
popularity_df  = num_rating_df.merge(avg_rating_df, on = 'Book-Title')
```

```python
popularity_df.merge(Books_df, on = 'Book-Title').drop_duplicates('Book-Title')
```

```python
popularity_df = popularity_df[popularity_df['Rating-Count']>=250].sort_values('Rating-Avg', ascending = False).head(50)
```

```python
popular_df = popularity_df.merge(Books_df, on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','Rating-Count','Rating-Avg']]
```

## collaborating filtering 

```python
ratings_with_name.head()
```

```python
# filter users who have rated min of 200 boook
x = ratings_with_name.groupby(['User-ID'])['Book-Rating'].count() >= 200
rating_over_200 = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(rating_over_200)]
```

```python
# filter books which have more than 50 ratings
y = filtered_rating.groupby(['Book-Title'])['Book-Rating'].count()>=50
```

```python
books_over_50 = y[y].index
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(books_over_50)]
```

```python
# contains users who have rated more than 200 books and books which have more than 50 ratings
final_ratings
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score[0]
```

```python
def recommend (book_name):
    # fetch index
    index = np.where(pt.index==book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])),key = lambda x:x[1], reverse = True)[1:11]
    
    data = []
    for i in similar_books:
        item=[]
        temp_df = Books_df[Books_df['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
         
        data.append(item)
    return data
```
